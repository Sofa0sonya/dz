```python
#3.1 Создайте с помощью функции-преобразования новый признак «Образование», который должен иметь четыре категории: «высшее», «неоконченное высшее», «среднее специальное» и «среднее».
#функция для отнесения уровня образования к заданным
def get_education(education):
    education = education.lower()
    education_ls = education.split(' ')
    if education_ls[0] == 'высшее':
        return 'высшее'
    elif education_ls[0] == 'неоконченное':
        return 'неоконченное высшее'
    elif education_ls[1] == 'специальное':
        return 'среднее специальное'
    elif education_ls[0] == 'среднее':
        return 'среднее'

#добавляем признак "Образование" и проверим количество уникальных категорий
df1 = df.copy()
df1['Образование'] = df1['Образование и ВУЗ'].apply(get_education)
print(f"Количество уникальных значений в признаке 'Образование': {df1['Образование'].nunique()}")
#удаляем старый признак "Образование и ВУЗ"
hh_data_copy = df1.drop(['Образование и ВУЗ'], axis=1)
#проверяем отсутствие пропусков в новом признаке
print(f'Количество пропусков в признаке "Образование": {df1["Образование"].isnull().sum()}')
#Сколько соискателей имеет средний уровень образования (школьное образование)?
print(f"Количество соискателей со средним уровнем образования: {df1['Образование'].value_counts().loc['среднее']}")
```

```python
#3.2 Создайте два новых признака «Пол» и «Возраст».
# новые признаки "Пол" и "Возраст" с помощью lambda функций
df1['Пол'] = df1['Пол, возраст'].apply(lambda x: x.split(',')[0][0])
df1['Возраст'] = df1['Пол, возраст'].apply(lambda x: int(x.split(',')[1][0:5]))
#удаляем старый признак "Пол, возраст"
df1 = df1.drop(['Пол, возраст'], axis=1)
#Сколько процентов женских резюме представлено в наших данных? Ответ округлите до сотых.
print(f"Резюме соискателей женского пола составляют: {round(df1['Пол'].value_counts(normalize=True)[1]*100, 2)}%")
#Чему равен средний возраст соискателей? Ответ округлите до десятых.
print(f"Средний возраст соискателей составляет: {round(df1['Возраст'].mean(), 1)} года")
```

```python
#3.3 В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
#функция для извлечения опыта работы в месяцах
def get_experience(str):
    year_ls = ['лет', 'года', 'год']
    month_ls = ['месяц','месяца', 'месяцев']
    year = 0
    month = 0
    if str == "Не указано" or str is np.nan: # Проверка на отсутствие данных
        return np.nan
    else:
        str = str.split(' ')
        for num in range(6):
            if str[num] in year_ls:
                year = int(str[num-1])
            elif str[num] in month_ls:
                month = int(str[num-1])
            else:
                continue
    return year*12 + month
#создаем признак 'Опыт работы (месяц)' и удалим старый признак
df1['Опыт работы (месяц)'] = df1['Опыт работы'].apply(get_experience)
df1 = df1.drop(['Опыт работы'], axis=1)
#Чему равен медианный опыт работы (в месяцах) в нашей таблице?
print(f"Медианное значение опыта работы: {df1['Опыт работы (месяц)'].median()} месяцев")
```

```python
#3.4 Создадим отдельные признаки «Город», «Готовность к переезду», «Готовность к командировкам».
#создаем функцию для определения города проживания соискателя
def get_city(str):
    million_cities = [
        'Новосибирск', 'Екатеринбург','Нижний Новгород',
        'Казань', 'Челябинск','Омск', 'Самара', 'Ростов-на-Дону',
        'Уфа', 'Красноярск', 'Пермь', 'Воронеж','Волгоград'
        ]
    str = str.split(' ')
    if str[0] == 'Москва':
        return 'Москва'
    elif str[0] == 'Санкт-Петербург':
        return 'Санкт-Петербург'
    elif str[0] in million_cities:
        return 'город-миллионник'
    else:
        return 'другие'

#создаем функцию для определения готовности к переезду
def get_trip(str):
    str = str.split(',')
    for x in str:
        if re.findall(r'ерее', x) == ['ерее']:
            if ('готов' in x or 'хочу' in x) and ('не готов' not in x):
                return True
            else:
                return False
#создаем функцию для определения готовности к командировкам
def get_business_trip(str):
    if 'командировка' in str:
        if (re.findall(r'не готов к командировкам', str) == ['не готов к командировкам']) \
        or (re.findall(r'не готова к командировкам', str) == ['не готова к командировкам']):
            return False
        else:
            return True
    else:
        return False

#создаем новые признаки и проверяем результат
df1['Город'] = df1['Город, переезд, командировки'].apply(get_city)
display(df1['Город'].value_counts())
df1['Готовность к переезду'] = df1['Город, переезд, командировки'].apply(get_trip)
display(df1['Готовность к переезду'].value_counts())
df1['Готовность к командировкам'] = df1['Город, переезд, командировки'].apply(get_business_trip)
display(df1['Готовность к командировкам'].value_counts())
#удаляем старый признак
df1 = df1.drop(['Город, переезд, командировки'], axis=1)
# Сколько процентов соискателей живут в Санкт-Петербурге? Ответ округлите до целого.
x = round(df1['Город'].value_counts(normalize=True)*100)['Санкт-Петербург']
print(f"В Санкт-Петербурге проживает {x}% соискателей")
# Сколько процентов соискателей готовы одновременно и к переездам, и к командировкам? Ответ округлите до целого.
mask = (df1['Готовность к командировкам']) & (df1['Готовность к переезду'])
x = round(df1[mask]['Готовность к командировкам'].count() / df1.shape[0]*100)
print(f"К переезду и командировкам готово {x}% соискателей")
```

```python
#3.5 Создание признаки-«мигалки» для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе — False.
#новые признаки с использованием lambda функций
df1['полная занятость'] = df1['Занятость'].apply(lambda x: True if 'полная занятость' in x else False)
df1['частичная занятость'] = df1['Занятость'].apply(lambda x: True if 'частичная занятость' in x else False)
df1['проектная работа'] = df1['Занятость'].apply(lambda x: True if 'проектная работа' in x else False)
df1['волонтерство'] = df1['Занятость'].apply(lambda x: True if 'волонтерство' in x else False)
df1['стажировка'] = df1['Занятость'].apply(lambda x: True if 'стажировка' in x else False)
df1['полный день'] = df1['График'].apply(lambda x: True if 'полный день' in x else False)
df1['сменный график'] = df1['График'].apply(lambda x: True if 'сменный график' in x else False)
df1['гибкий график'] = df1['График'].apply(lambda x: True if 'гибкий график' in x else False)
df1['удаленная работа'] = df1['График'].apply(lambda x: True if 'удаленная работа' in x else False)
df1['вахтовый метод'] = df1['График'].apply(lambda x: True if 'вахтовый метод' in x else False)
#удаляем старый признак
df1 = df1.drop(['Занятость', 'График'], axis=1)
#Сколько людей ищут проектную работу и волонтёрство (в обоих столбцах стоит True)?
mask1 = (df1['проектная работа'] == True) & (df1['волонтерство'] == True)
print(f'Количество соискателей ищущих проектную работу и волонтёрство: {df1[mask1].shape[0]} человек')
#Сколько людей хотят работать вахтовым методом и с гибким графиком (в обоих столбцах стоит True)?
mask2 = (df1['вахтовый метод'] == True) & (df1['гибкий график'] == True)
print(f'Количество соискателей, которые хотят работать вахтовым методом и с гибким графиком: {df1[mask2].shape[0]} человек')
```

```python
#3.6 Создайте новый DataFrame из скачанного файла
hh_data_copy['Обновление резюме'] = pd.to_datetime(hh_data_copy['Обновление резюме'], dayfirst=True)
hh_data_copy['Обновление резюме'] = hh_data_copy['Обновление резюме'].dt.date
#считываем таблицу с валютами и преобразуем дату в формат datetime
rates_df = pd.read_csv(
    'https://github.com/Sofa0sonya/dz/blob/main/ExchangeRates.zip',
    compression='zip',
    usecols=[
        "currency",
        "date",
        "proportion",
        "close"
        ]
    )
rates_df['date'] = pd.to_datetime(rates_df['date'], dayfirst=True)
#создаем новые признаки с суммой желаемой зарплаты в числовом формате и наименование валюты
df1['currency'] = df1['ЗП'].apply(lambda x: x.split()[1])
df1['sum'] = df1['ЗП'].apply(lambda x: x.split()[0])
df1['sum'] = df1['sum'].astype('float')
#Наименование валюты переводим в стандарт ISO
cur_dict ={
    'руб.':'RUB',
    'грн.':'UAH',
    'USD':'USD',
    'EUR':'EUR',
    'бел.руб.':'BYN',
    'KGS':'KGS',
    'сум':'UZS',
    'AZN':'AZN',
    'KZT':'KZT'
    }
df1['currency'] = df1['currency'].apply(lambda x: cur_dict[x])
#объединяем две таблицы по признаку валюты и дате
hh_df = df1.copy()
hh_df['Обновление резюме'] = pd.to_datetime(hh_df['Обновление резюме'])
hh_df = hh_df.merge(
    rates_df,
    left_on=['currency', 'Обновление резюме'],
    right_on=['currency', 'date'],
    how='left'
    )
#заполняем отсутствующие значения close и proportion единицей для рубля
hh_df['close'] = hh_df['close'].fillna(1)
hh_df['proportion'] = hh_df['proportion'].fillna(1)
#создаем новый признак, содержащий желаемую зарплату в рублях
hh_df['ЗП (руб)'] = (hh_df['sum'] * hh_df['close']) / hh_df['proportion']
#удаляем вспомагательные признаки
hh_df = hh_df.drop(['ЗП','currency', 'sum', 'date', 'close', 'proportion'], axis=1)
#Чему равна желаемая медианная заработная плата соискателей в нашей таблице (в рублях)?
x = round(hh_df['ЗП (руб)'].median() / 1000)
print(f"Медианное значение заработной платы составляет {x} тысяч рублей.")
```

```python
#4.1 Постройте распределение признака "Возраст"
#создаем гистограмму
fig = px.histogram(
    data_frame=hh_df,
    x='Возраст',
    marginal='box',
    width=800,
    height=500,
    )
fig.update_layout(
    title='Распределение возраста соискателей',
    title_x = 0.5,
    xaxis_title='Возраст соискателей, год',
    yaxis_title='Число наблюдений'
    )
fig.show()
fig = px.histogram(
    data_frame=hh_df,
    x=np.log(hh_df['Возраст']),
    marginal='box',
    width=800,
    height=500,
    )
fig.add_shape(
    type='line',
    line_color='red',
    line_width=3,
    opacity=1,
    line_dash='solid',
    x0=np.log(hh_df['Возраст']).mean(),
    x1=np.log(hh_df['Возраст']).mean(),
    xref='x',
    y0=0,
    y1=3000,
    yref='y'
    )
fig.update_layout(
    title='Распределение возраста соискателей Log',
    title_x = 0.5,
    xaxis_title='Возраст соискателей',
    yaxis_title='Число наблюдений'
    )
fig.show()
#создаем диаграмму размаха для признака возраст
fig, axes = plt.subplots(2, 2, figsize=(12, 12))
bx = sns.boxplot(
    data=hh_df,
    x='Возраст',
    orient='h',
    ax=axes[0][0]
    )
bx.set_title('Диаграмма размаха возраста соискателей')
bx.set_xlabel('возраст соискателя, год')
bx.grid()
hist = sns.histplot(
    data=hh_df,
    x='Возраст',
    ax=axes[1][0])
hist.set_title('Гистограмма распределения возраста соискателей')
hist.set_xlabel('Возраст соискателя, год')
hist.set_ylabel('Количество наблюдений')
hist.grid()
bx = sns.boxplot(
    data=hh_df,
    x=np.log(hh_df['Возраст']),
    orient='h',
    ax=axes[0][1]
    )
bx.set_title('Диаграмма размаха возраста соискателей Log')
bx.set_xlabel('возраст соискателя')
bx.grid()
hist = sns.histplot(
    data=hh_df,
    x=np.log(hh_df['Возраст']),
    ax=axes[1][1])
hist.axvline(np.log(hh_df['Возраст']).mean(), color='r', lw='2')
hist.set_title('Гистограмма возраста соискателей Log')
hist.set_xlabel('Возраст соискателя')
hist.set_ylabel('Количество наблюдений')
hist.grid()
plt.show()
print('Показатель ассиметрии:', round(np.log(hh_df['Возраст']).skew(), 2))
```
